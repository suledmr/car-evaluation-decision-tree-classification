```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("/kaggle/input/datasets/elikplim/car-evaluation-data-set/car_evaluation.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
col_names = ["buying","maint", "doors","people","lug_boot","safety","class"]
```

```python
df.columns = col_names
```

```python
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    print(df[col].value_counts())
```

```python
#class -> target variable, doors,people -> numeric, rest->categoric
```

```python
df.isnull().sum()
```

```python
#Feature Engineering
```

```python
df["doors"].unique()
```

```python
df["doors"] = df["doors"].replace("5more","5")
df["doors"].unique()
```

```python
df["doors"] = df["doors"].astype(int)
```

```python
df["people"].unique()
```

```python
df["people"] = df["people"].replace("more","5")
df["people"].unique()
```

```python
df["people"] = df["people"].astype(int)
```

```python
X = df.drop("class", axis = 1)
y = df["class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
#Encoding -> ordinal encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
categorical_cols = ["buying","maint","lug_boot","safety"]
numerical_cols = ["doors","people"]

ordinal_encoder = OrdinalEncoder(categories=[
    ["low","med","high","vhigh"], #buying
    ["low","med","high","vhigh"], #maint
    ["small","med","big"], #lug_boot
    ["low","med","high"] #safety
])

preprocessor = ColumnTransformer(transformers= [
    ("transformation_name", ordinal_encoder, categorical_cols),
], remainder = "passthrough")

X_train_transform = preprocessor.fit_transform(X_train)
X_test_transform = preprocessor.transform(X_test)
```

```python
pd.DataFrame(X_train_transform)
```

```python
#Train with Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)

tree_model.fit(X_train_transform,y_train)
```

```python
y_pred = tree_model.predict(X_test_transform)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score, classification_report 
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
plt.figure(figsize=(12,8))

from sklearn import tree

column_names = categorical_cols + numerical_cols
tree.plot_tree(tree_model.fit(X_train_transform,y_train),feature_names=column_names)
```

```python
# hyperparameter tuning
```

```python
param = {
    "criterion" : ["gini","entropy","log_loss"],
    "splitter" : ["best","random"],
    "max_depth" : [1,2,3,4,5,15,None],
    "max_features" : ["sqrt","log2",None]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=DecisionTreeClassifier(),param_grid=param,cv=5,scoring="accuracy")
```

```python
import warnings
warnings.filterwarnings("ignore")
grid.fit(X_train_transform,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test_transform)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
tree_model_best = DecisionTreeClassifier(criterion="entropy",max_depth=None,max_features=None,splitter="best")
```

```python
tree_model_best.fit(X_train_transform,y_train)
y_pred = tree_model_best.predict(X_test_transform)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
tree.plot_tree(tree_model_best.fit(X_train_transform,y_train),feature_names=column_names)
```
